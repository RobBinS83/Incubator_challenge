## In 2017, what was the average number of beneficiaries per provider? Due to the suppression of data for those with few beneficiaries, we can only include those with more than 10 beneficiaries.

```python
num_provider = rawdata["bene_count"].apply(lambda x: x > 10)

cleaned = rawdata["bene_count"][num_provider]
print("Average number of beneficiaries per provider is {:.1f}".format(cleaned.sum()/len(cleaned)))
```

## Work out for each Specialty the fraction of drug claims that are for brand-name drugs. Include only providers for whom the relevant information has not been suppressed, and consider only specialties with at least 1000 total claims. What is the standard deviation of these fractions?

```python
set(rawdata["specialty_description"])
```

```python
selected_table = pd.DataFrame(0, index = (0, 1), columns = set(rawdata["specialty_description"]))

m  = len(rawdata)
tic = time.time()

for i in range(m):
    a = rawdata["brand_claim_count"][i]
    b = rawdata["total_claim_count"][i]
    if  a > 0 and  b > 0:
        selected_table.loc[0,rawdata["specialty_description"][i]] += a
        selected_table.loc[1,rawdata["specialty_description"][i]] += b
        
toc = time.time()
print("Processing {:d} cases takes {:f} seconds.".format(m, toc-tic))
```

```python
selected_table = selected_table.transpose()
selected_table.head(20)
```

## Let's find which states have surprisingly high supply of opioids, conditioned on specialty. Work out the average length of an opioid prescription for each provider. For each (state, specialty) pair with at least 100 providers, calculate the average of this value across all providers. Then find the ratio of this value to an equivalent quantity calculated from providers in each specialty across all states. What is the largest such ratio?

```python
specialty_list = set(rawdata["specialty_description"])
state_list = set(rawdata["nppes_provider_state"])

res = pd.DataFrame(0, index = state_list, columns = specialty_list)

m = len(rawdata)

tic = time.time()

for i in range(m):
    a = rawdata["opioid_day_supply"][i]
    if  a > 0:
        res.loc[rawdata["nppes_provider_state"][i],rawdata["specialty_description"][i]] += a
        
toc = time.time()
print("Processing {:d} cases takes {:f} seconds.".format(m, toc-tic))
```

```python
for item in specialty_list:
    print("{:s}: {:10s}".format(item, res[item].idxmax()))
```

```python
specialty_list = set(rawdata["specialty_description"])
state_list = set(rawdata["nppes_provider_state"])

res1 = pd.DataFrame(0, index = state_list, columns = specialty_list)
res2 = pd.DataFrame(0, index = state_list, columns = specialty_list)

m = len(rawdata)

tic = time.time()

for i in range(m):
    a = rawdata["opioid_day_supply"][i]
    b = rawdata["opioid_claim_count"][i]
    if  b > 0:
        res1.loc[rawdata["nppes_provider_state"][i],rawdata["specialty_description"][i]] += 1.0 * a / b
        res2.loc[rawdata["nppes_provider_state"][i],rawdata["specialty_description"][i]] += 1.0
        
toc = time.time()
print("Processing {:d} cases takes {:f} seconds.".format(m, toc-tic))
```

## For each provider, estimate the length of the average prescription from the total_day_supply and total_claim_count. What is the median, in days, of the distribution of this value across all providers?

```python
res = pd.DataFrame(0, index = range(len(rawdata)), columns = ["average_prescription"])
m = len(rawdata)
#m = 10000
tic = time.time()
for i in range(m):
    res.loc[i] = rawdata.loc[i, "total_day_supply"] / rawdata.loc[i, "total_claim_count"] 
    
toc = time.time()
print("Processing {:d} cases takes {:f} seconds.".format(m, toc-tic))
```

```python
res.median()
```

## Find the ratio of beneficiaries with opioid prescriptions to beneficiaries with antibiotics prescriptions in each state. Assume that each beneficiary attends only a single provider. What is the difference between the largest and smallest ratios?

```python
state_list = set(rawdata["nppes_provider_state"])
res = pd.DataFrame(0, index = range(3), columns = state_list)

m = len(rawdata)
#m = 10000
tic = time.time()
for i in range(m):
    if rawdata.loc[i, "opioid_bene_count"] > 0 and rawdata.loc[i, "antibiotic_bene_count"] > 0:
        res.loc[0, rawdata.loc[i, "nppes_provider_state"]] += rawdata.loc[i, "opioid_bene_count"]
        res.loc[1, rawdata.loc[i, "nppes_provider_state"]] += rawdata.loc[i, "antibiotic_bene_count"]
    
toc = time.time()
print("Processing {:d} cases takes {:f} seconds.".format(m, toc-tic))
```

## For each provider where the relevant columns are not suppressed, work out the fraction of claims for beneficiaries age 65 and older, as well as the fraction of claims for beneficiaries with a low-income subsidy. What is the Pearson correlation coefficient between these values?

```python
res = pd.DataFrame(0, index = range(len(rawdata)), columns = ["ge65", "lis"])

for i in range(len(rawdata)):
    res.loc[i,"ge65"] = rawdata.loc[i, "total_claim_count_ge65"] / rawdata.loc[i, "total_claim_count"]
    res.loc[i,"lis"] = rawdata.loc[i, "lis_claim_count"] / rawdata.loc[i, "total_claim_count"]
    
```

```python
res.dropna().shape
```

```python
coef = res.dropna().corr()
print(coef)
```
